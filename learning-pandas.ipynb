## Series
Each column in a DataFrame is a Series

```python
s1 = pd.Series(['a', 'b', 'c', 'd'])
s2 = pd.Series(['e', 'f', 'g', 'h'])
```

it's possible to change the index, and give a name:

```python
pd.Series(['a', 'b', 'c', 'd'], index=[1,2,3,4], name='Random')
```
